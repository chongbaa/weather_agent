```python
from openai import OpenAI
from dotenv import load_dotenv

# 加载 .env 文件
load_dotenv()

# 直接创建客户端，它会自动读取 OPENAI_API_KEY
client = OpenAI()

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", "content": "用中文自我介绍一下"}
    ]
)

print(completion.choices[0].message.content)
```
